空值和缺失值的处理

```python
import pandas as pd
from numpy import NaN
data_obj=pd.Series([1,None,NaN])
data_obj
```

```python
# isnull检查一下是否有空值或缺失值
pd.isnull(data_obj)
```

```python
#notnull 检查一下是否不为空值或缺失值
pd.notnull(data_obj)
```

```python
import pandas as pd
import numpy as np
df_obj = pd.DataFrame({"类别":['小说','散文随笔','青春文学','传记'],
                       "书名":[np.nan,'《皮囊》','《旅程结束时》','《老舍自传》'],
                       "作者":["老舍",None,"张启鑫","老舍"]})
df_obj
```

```python
#删除数据集中空值和缺失值
df_obj.dropna()
```

```python
#使用fillna进行填充
df_obj = pd.DataFrame({'A':[1,2,3,NaN],
                       'B':[NaN,4,NaN,6],
                       'C':['a',7,8,9],
                       'D':[NaN,2,3,NaN]})
df_obj
```

```python
df_obj.fillna("66.0")#使用的是66.0进行数据替换缺失值
```

```python
#对单个数据进行填充
df_obj.fillna({'A':4.0,'B':5.0})
```

```python
#重复值的处理
person_info = pd.DataFrame({'id':[1,2,3,4,4,5],
                            'name':['小铭','小月月','彭岩','刘华','刘华','周华'],
                            'age':[18,18,29,58,58,36],
                            'height':[180,180,185,175,175,178],
                            'gender':['女','女','男','男','男','男']})
person_info
```

```python
#我像判断数据里面是不是有重复值
person_info.duplicated()
```

```python
#把重复值进行删除
person_info.drop_duplicates()
```

```python
person_info
```

```python
person_info.drop_duplicates(inplace=True)
```

```python
person_info
```

```python
#数据合并
import pandas as pd
df1 = pd.DataFrame({'A':['A0','A0','A1'],
                    'B':['B0','B0','B1'],
                    'C':['C0','C0','C1']})
df1
```

```python
df2 = pd.DataFrame({'B':['B0','B0','B1'],
                    'C':['C0','C0','C1'],
                    'D':['D0','D2','D2']})
df2
```

```python
#数据合并
#outer 是外连接 并集
#inner 是内连接 交际
pd.concat([df1,df2],join='outer',axis=1)
```

```python
pd.concat([df1,df2],join='inner',axis=0)
```

数据的应用

```python
data_frame = pd.DataFrame({'data1':[80,23,25,63,94,92,99,92,82,99],
                           'data2':[41,87,58,68,72,89,60,42,53,65],
                           'data3':[30,78,23,66,16,59,20,23,24,40],
                           'key':list('baabbabaaa')})
data_frame
```

```python
#groupby 方法进行分组
data = data_frame.groupby("key")
for x in data:
    print(x)
```

```python
dict([x for x in data])["a"]
```

```python
dict([x for x in data])["b"]
```

```python
data.apply(lambda x:x.sum()-x.min())
```
