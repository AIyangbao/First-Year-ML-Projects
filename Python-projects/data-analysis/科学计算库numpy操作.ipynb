```python
a1=np.arange(1,7).reshape(2,3)
a1
```

```python
a2=np.arange(7,13).reshape(2,3)
a2
```

```python
#我想把两个数据进行合并 #默认axis=0
np.concatenate((a1,a2))
```

```python
#默认axis=0 按行
np.concatenate((a1,a2))
```

```python
#axis=1 按列
np.concatenate((a1,a2),axis=1)
```
