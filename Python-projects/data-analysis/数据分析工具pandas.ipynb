```python
#Pandas索引操作以及高级索引
ser_obj=pd.Series([1,2,3,4,5],index=['c','d','a','b','e'])
ser_obj
```

```python
#重新索引
ser_obj2=ser_obj.reindex(['a','b','c','d','e','f'])
ser_obj2
```

```python
ser_obj2=ser_obj.reindex(['a','b','d','e','f'],fill_value=6)
ser_obj2
```

```python
#DataFrame.reindex(index=None,colums=None,method=None,copy=True,Fill_value=nan,limit=None)
ser_obj3=pd.Series([1,3,5,7],index=[0,2,4,6])
ser_obj3
```

```python
ser_obj3.reindex(range(6),method='ffill')##重新索引，向前填充值
```

```python
for i in range(6):
    print(i)
```

```python
ser_obj3.reindex(range(6),method='bfill')## #重新索引，向后填充值
```
