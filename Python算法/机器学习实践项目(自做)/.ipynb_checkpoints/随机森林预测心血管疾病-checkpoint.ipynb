```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree
```

```python
df= pd.read_csv('心血管疾病.csv')
df
```

```python
#相关性热图
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei']#用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),cmap='coolwarm')
```

```python
#当两个特征相关系数比较大 那么训练的时候 就可以只采取一个
X=df.drop(columns=['是否吸烟','是否死亡'])
y=df['是否死亡']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = RandomForestClassifier(n_estimators=300)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
import numpy as np
def predict_cardiovascular_disease(年龄,贫血,肌酸酐磷酸激酶,糖尿病,射血分数,高血压,血小板,血清肌酐,血清钠,性别,是否吸烟,随访次数):
  input_features=np.array([[年龄,贫血,肌酸酐磷酸激酶,糖尿病,射血分数,高血压,血小板,血清肌酐,血清钠,性别,随访次数]])
  predict_result=model.predict(input_features)
  if predict_result[0]==0:
      return "存活"
  else:
      return"死亡"
  
```
