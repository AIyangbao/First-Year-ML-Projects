```python
#1.鸢尾花数据集
from sklearn.datasets import load_iris
#导入画图的库
import matplotlib.pyplot as plt
```

```python
load_iris()
```

```python
iris=load_iris()
X=iris.data#输入数据
Y=iris.target#标签
```

```python
feature_names=iris.feature_names
target_names=iris.target_names
```

```python
target_names
```

```python
#绘制散点图
#配置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']#设置全局字体为黑体
plt.rcParams['axes.unicode_minus'] = False#解决负号显示异常
```

```python
#设置画布大小
plt.figure(figsize=(8,6))
#散点图 方法：scatter 萼片长度 VS 萼片宽度
#X是一个numpy 一种结果 取值[行索引，列索引]
plt.scatter(X[0,0],X[0,1])
```

```python
#总共3个类别
for i in range(3):
    plt.scatter(X[Y==i,0],X[Y==i,1])
plt.xlabel("萼片长度")
plt.ylabel("萼片宽度")
plt.title("鸢尾花数据集")
```

```python
for i in range(3):
    plt.scatter(X[Y==i,2],X[Y==i,3])
plt.xlabel("花瓣长度")
plt.ylabel("花瓣宽度")
plt.title("鸢尾花数据集")
```

```python
#批量化一下数字图片
plt.figure(figsize=(10,4))
target=digit.target
for i in range(16):
    plt.subplot(2,8,i+1)
    plt.imshow(images[i])
    plt.title(target[i])
```
